Создаем массив из средних выборочных, строим по нему гистограмму и сравниваем с плотностью нормального распределения (по теоретическим параметрам):

### Объём выборки: n=2

```python
massiv2 = [] #переменная массива
n=2 #объём выборки

#цикл для наполнения массива средними значениями выборки
for i in range (0,1000):
    sample2 = beta_rv.rvs(n) #создаем выборку из n элементов
    sum = np.sum(sample2)/n   # считаем среднее
    massiv2.append(sum) #добавляем элемент в массив

#строим гистограмму
plt.hist(massiv2, bins=10, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

#считаем среднее, дисперсию, стандартное отклонение
srednee= alfa/(alfa+beta)
dispersia = (alfa*beta)/((alfa+beta)**2*(alfa+beta+1))
otkl = math.sqrt((n/(n-1))*dispersia**2.)
#otkl = math.sqrt(dispersia)
#задаем функцию распределения
norm_rv2 = sts.norm(loc=srednee, scale=otkl)

#строим график нормального распределения
x2 = np.linspace(0,1,100)
pdf2 = norm_rv2.pdf(x2)
plt.plot(x2,pdf2)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
